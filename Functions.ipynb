1. What is a difference between function and method in python
 -  Function is an independent block of code that performs a task and is not    bound to any object. Functions are defined using the def keyword and can be called by name, optionally with arguments.
 -  A method, by contrast, is a function that is associated with an object. Methods are defined within a class and are called using the dot (.) notation on an object instance. They typically take self as the first parameter, referring to the object itself.

6. Explain the concept of generators in Python and how they are defined?
 - A generator is a special type of function that returns an iterator. Instead of using the return keyword, generators use the yield keyword to produce a series of values one at a time, pausing after each yield and resuming from where they left off.

```
def even_numbers(n):
    for i in range(n+1):
        if i % 2 == 0:
            yield i

for num in even_numbers(6):
    print(num) # Output: 0,2,4,6
```



7. What are the advantages of using generators over regular functions?
- Advantages of Generators Over Regular Functions in Python:
 - Memory Efficient: Generate values one at a time without storing the entire result in memory.

 - Lazy Evaluation: Values are computed only when needed using yield.

 - Handle Infinite Sequences: Can produce endless data (e.g., infinite counters) safely.

 - Simpler Syntax: Easier to write than custom iterator classes.

 - Faster Execution: Better performance in loops over large datasets.

 - Support Data Pipelines: Can be chained for step-by-step data processing.

9. Explain the purpose and usage of the `map()` function in Python.
 - The map() function applies a given function to each item of an iterable (like a list, tuple, etc.).

   It returns a map object, which is an iterator that can be converted into a list or other sequence.

      ```
      numbers = [1, 2, 3, 4]
      squares = map(lambda x: x * x, numbers)
      print(list(squares))  # Output: [1, 4, 9, 16]
      ```


10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
 - In Python, map(), filter(), and reduce() are built-in functions used for processing iterables in a functional programming style.  

   -   map() is used to apply a function to each item in an iterable and returns a new iterable with the results.

    -  filter() selects elements from an iterable for which a given function returns True, effectively filtering the data.

    - reduce() repeatedly applies a function to the items of an iterable, reducing it to a single cumulative value.


```
numbers = [1, 2, 3, 4]
result = reduce(lambda x, y: x + y, numbers)
print(result)  # Output: 10
```

````python
#  Python function that checks if a given number is prime or not from 1 to 200.
def prime(num):
  if num < 2:
    return False
  for i in range(2, int(num**0.5+1)):
    if num % i == 0:
      return False
  return True

print(prime(17))
print(prime(23))
print(prime(24))
````

````python
# Create an iterator function in Python that generates the Fibonacci sequence up to a specified number of terms.
def fibo_gen(numb) :
  a,b= 0,1;
  for i in range(numb):
    yield a
    a,b = b,a+b

for i in fibo_gen(10):
  print(i)

fib = fibo_gen(10)
print(next(fib))
print(next(fib))
print(next(fib))
print(next(fib))
print(next(fib))
````

````python
# Write a generator function in Python that yields the powers of 2 up to a given exponent.
def exp(exponent):
  for i in range(exponent+1):
    yield 2**i

for i in exp(10):
  print(i)
````
